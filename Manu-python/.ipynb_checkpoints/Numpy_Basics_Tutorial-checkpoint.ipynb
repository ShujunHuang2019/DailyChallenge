## N-Dimensional array

```python
import numpy as np
scores = [85, 56, 30, 45, 69]
first_arr = np.array(scores)
print first_arr
print first_arr.dtype
```

```python
scores_1 = [[2, 4, 5], [5, 2, 5]]
second_arr = np.array(scores_1)
print second_arr
print second_arr.ndim
print second_arr.shape
print second_arr.dtype
```

```python
x = np.zeros(10)
x = np.ones(10)
print x
print len(x)
```

```python
x = np.zeros((4, 2))
print x
```

```python
print np.arange(15)
print range(15)
print np.eye(6)
```

```python
print first_arr
print first_arr * first_arr
print first_arr - first_arr
print 1/(first_arr)
print first_arr ** 0.5
```

```python
new_arr = np.arange(12)
print new_arr
print new_arr[5]
print new_arr[4:9]
new_arr[4:9] = 99
print new_arr
print new_arr[1]
```

```python
modi_arr = new_arr[4:9]
modi_arr[1] = 12345
print new_arr
modi_arr[:]
```

```python
matrix_arr = np.array([[1, 3, 4], [2, 3, 4], [4, 5, 5], [4, 8, 9]])
print matrix_arr
print matrix_arr[1]
print matrix_arr[0][1]
print matrix_arr[0, 1]
```
